```python
from sklearn.model_selection import train_test_split
```

```python
X= df.drop(['target'],1)
y=df.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=50).fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.svm import LinearSVC
svm = LinearSVC(C=50).fit(X_train,y_train)
```

```python
svm.score(X_test,y_test)
```

```python
import lightgbm as lgb
from sklearn.model_selection import RandomizedSearchCV,KFold


dtrain = lgb.Dataset(X_train,y_train)
dtest = lgb.Dataset(X_test,y_test)
params = {
        'boosting_type': ['dart','goss','gbdt'],    
        'learning_rate': [0.33,0.01,0.1],      
        'max_depth': [5,8,15,6],             
        'min_data_in_leaf': [4],      
        'max_bin' : [100,1000],   
        'num_iterations': [100,1000,10000],
        'num_leaves' : [128,100,50],        
        'num_iteration': [100] 
        
}


model_lgb = lgb.LGBMClassifier(objective="binary",metric="binary_error")
gkf = KFold(n_splits=5, shuffle=True, random_state=42).split(X=X, y=y)


from sklearn.metrics import make_scorer,roc_auc_score
scorer = make_scorer(roc_auc_score)

grid = RandomizedSearchCV(model_lgb,params,n_iter=100,cv=gkf,scoring=scorer).fit(X,y)
```

```python
import xgboost as xgb

xgb_params = {
        'booster': ['gbtree','gblinear'],    
        'eta': [0.33,0.01,0.1],      
        'max_depth': [5,8,15,6],             
        'min_data_in_leaf': [4],      
        'max_bin' : [100,1000],   
        'num_iterations': [100,1000,10000],
        'num_leaves' : [128,100,50],        
        'num_iteration': [100] 
        
}


model_xgb = xgb.XGBClassifier(objective="binary:logistic",eval_metric="auc")
gkf = KFold(n_splits=5, shuffle=True, random_state=42).split(X=X, y=y)


from sklearn.metrics import make_scorer,roc_auc_score
scorer = make_scorer(roc_auc_score)

grid_xgb = RandomizedSearchCV(model_xgb,xgb_params,n_iter=100,cv=gkf,scoring=scorer).fit(X,y)
```
